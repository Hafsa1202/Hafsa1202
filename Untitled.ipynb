```python
ct=(df["category_name_1"] == "Men's Fashion").sum()
print("sales:",ct)
```

```python
product_sales=df.groupby('sku')['qty_ordered'].sum().reset_index()
product_sales=product_sales.sort_values(by='qty_ordered',ascending=False)
top_product=product_sales.head(3)
print(top_product)
```

```python
dt=df.groupby('grand_total') ['Month'].sum()
print(dt)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(7,4))
ax=sns.countplot(x='Month' , data=df)
for bars in  ax.containers:
    ax.bar_label(bars)
```
